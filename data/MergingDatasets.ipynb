```python
import pandas as pd
import re
import numpy as np
```

```python
#activity = pd.read_csv('pos_state_collar_2_2018_08_16_precise.csv')
collarC3 = pd.read_csv('9dof_processed_C3.csv')
braceletC3 = pd.read_csv('ble_processed_C3.csv')
gpsC3 = pd.read_csv('gps_processed_C3.csv')
gpsC3['TIME']=gpsC3['TIMESTAMP']
del gpsC3['TIMESTAMP']
collarC3.columns=['TIME', 'collar_ACC_X', 'collar_ACC_Y', 'collar_ACC_Z', 'collar_GYRO_X', 'collar_GYRO_Y', 'collar_GYRO_Z',
       'collar_MAG_X', 'collar_MAG_Y', 'collar_MAG_Z']
braceletC3.columns=['TIME', 'bracelet_ACC_X', 'bracelet_ACC_Y', 'bracelet_ACC_Z', 'bracelet_GYRO_X', 'bracelet_GYRO_Y', 'bracelet_GYRO_Z']
```

```python
collarC2 = pd.read_csv('9dof_processed_C2.csv')
braceletC2 = pd.read_csv('ble_processed_C2.csv')
gpsC2 = pd.read_csv('gps_processed_C2.csv')
gpsC2['TIME']=gpsC2['TIMESTAMP']
del gpsC2['TIMESTAMP']
collarC2.columns=['TIME', 'collar_ACC_X', 'collar_ACC_Y', 'collar_ACC_Z', 'collar_GYRO_X', 'collar_GYRO_Y', 'collar_GYRO_Z',
       'collar_MAG_X', 'collar_MAG_Y', 'collar_MAG_Z']
braceletC2.columns=['TIME', 'bracelet_ACC_X', 'bracelet_ACC_Y', 'bracelet_ACC_Z', 'bracelet_GYRO_X', 'bracelet_GYRO_Y', 'bracelet_GYRO_Z']
```

```python
min_time=min(collarC2['TIME'])
if min_time > min(braceletC2['TIME']):
    min_time=min(braceletC2['TIME'])
if min_time > min(gpsC2['TIME']):
    min_time=min(gpsC2['TIME'])
if min_time > min(collarC3['TIME']):
    min_time=min(collarC3['TIME'])
if min_time > min(braceletC3['TIME']):
    min_time=min(braceletC3['TIME'])
if min_time > min(gpsC3['TIME']):
    min_time=min(gpsC3['TIME'])
```

```python
max_time=max(collarC2['TIME'])
if max_time < max(braceletC2['TIME']):
    max_time=max(braceletC2['TIME'])
if max_time < max(gpsC2['TIME']):
    max_time=max(gpsC2['TIME'])
if max_time < max(collarC3['TIME']):
    max_time=max(collarC3['TIME'])
if max_time < max(braceletC3['TIME']):
    max_time=max(braceletC3['TIME'])
if max_time < max(gpsC3['TIME']):
    max_time=max(gpsC3['TIME'])
```

```python
time_list=[]
for i in range(min_time,max_time+1):
    time_list.append(i)
time_list=pd.DataFrame(time_list)
time_list.columns=['TIME']
```

```python
collarC3=collarC3.groupby(['TIME'],group_keys=time_list)[ 'collar_ACC_X', 'collar_ACC_Y', 'collar_ACC_Z', 'collar_GYRO_X', 'collar_GYRO_Y', 'collar_GYRO_Z',
       'collar_MAG_X', 'collar_MAG_Y', 'collar_MAG_Z'].mean()
braceletC3=braceletC3.groupby(['TIME'],group_keys=time_list)['bracelet_ACC_X', 'bracelet_ACC_Y', 'bracelet_ACC_Z', 'bracelet_GYRO_X', 'bracelet_GYRO_Y', 'bracelet_GYRO_Z'].mean()
gpsC3=gpsC3.groupby(['TIME']).mean()
```

```python
collarC2=collarC2.groupby(['TIME'],group_keys=time_list)[ 'collar_ACC_X', 'collar_ACC_Y', 'collar_ACC_Z', 'collar_GYRO_X', 'collar_GYRO_Y', 'collar_GYRO_Z',
       'collar_MAG_X', 'collar_MAG_Y', 'collar_MAG_Z'].mean()
braceletC2=braceletC2.groupby(['TIME'],group_keys=time_list)['bracelet_ACC_X', 'bracelet_ACC_Y', 'bracelet_ACC_Z', 'bracelet_GYRO_X', 'bracelet_GYRO_Y', 'bracelet_GYRO_Z'].mean()
gpsC2=gpsC2.groupby(['TIME'],group_keys=time_list).mean()
```

```python
mergedc3=pd.merge(gpsC3,pd.merge(braceletC3,collarC3,how='outer',on="TIME"),how='outer',on="TIME")
mergedc2=pd.merge(gpsC2,pd.merge(braceletC2,collarC2,how='outer',on="TIME"),how='outer',on="TIME")
```

```python
mergedc3=pd.merge(time_list,mergedc3,how='outer',on="TIME")
mergedc2=pd.merge(time_list,mergedc2,how='outer',on="TIME")
```

```python
actc3=pd.read_csv('activityc3.csv')
actc2=pd.read_csv('activityc2.csv')
actc3=actc3.dropna()
actc2=actc2.dropna()
actc3['start'] = actc3['start.time'].apply(lambda x: '{:.0f}'.format(x/1000))
actc3['end'] = actc3['end.times'].apply(lambda x: '{:.0f}'.format(x/1000))
actc2['start'] = actc2['start.time'].apply(lambda x: '{:.0f}'.format(x/1000))
actc2['end'] = actc2['end.times'].apply(lambda x: '{:.0f}'.format(x/1000))
del actc2['end.times']
del actc3['end.times']
del actc2['start.time']
del actc3['start.time']
```

```python
timestamps=[]
activity=[]
for row in range(len(actc2)):
    #print(actc2['start'][row])
    for i in range(int(actc2['start'][row]),int(actc2['end'][row])):
        #print (i)
        timestamps.append(i)
        activity.append(actc2['Behavior'][row])
activity

activityc2 = pd.DataFrame(
    {'TIME': timestamps,
     'activity': activity   
    })
```
